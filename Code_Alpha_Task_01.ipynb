```python
from flask import Flask, request, render_template_string
from flask_ngrok import run_with_ngrok

app = Flask(__name__)
run_with_ngrok(app)  # Starts ngrok when the app is run

html_template = '''
<!doctype html>
<html lang="en">
  <head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1, shrink-to-fit=no">
    <title>Fibonacci Sequence Generator</title>
    <link href="https://fonts.googleapis.com/css2?family=Roboto:wght@400;500;700&display=swap" rel="stylesheet">
    <style>
      body {
        font-family: 'Roboto', sans-serif;
        background-color: #f8f9fa;
        color: #343a40;
        margin: 0;
        padding: 0;
      }
      .container {
        max-width: 600px;
        margin: 50px auto;
        background: #ffffff;
        padding: 20px 30px;
        box-shadow: 0 0 10px rgba(0, 0, 0, 0.1);
        border-radius: 8px;
      }
      h1 {
        text-align: center;
        color: #007bff;
        margin-bottom: 30px;
      }
      form {
        text-align: center;
      }
      label {
        font-weight: 500;
        margin-bottom: 10px;
        display: inline-block;
      }
      input[type="number"] {
        padding: 10px;
        width: calc(100% - 22px);
        margin-bottom: 20px;
        border: 1px solid #ced4da;
        border-radius: 4px;
      }
      button {
        padding: 10px 20px;
        background-color: #007bff;
        color: #ffffff;
        border: none;
        border-radius: 4px;
        cursor: pointer;
        font-size: 16px;
      }
      button:hover {
        background-color: #0056b3;
      }
      .result {
        margin-top: 20px;
        text-align: center;
        background: #e9ecef;
        padding: 15px;
        border-radius: 4px;
      }
      .result h2 {
        margin: 0 0 10px;
        color: #495057;
      }
      .result p {
        margin: 0;
        font-size: 18px;
        color: #212529;
      }
    </style>
  </head>
  <body>
    <div class="container">
      <h1>Fibonacci Sequence Generator</h1>
      <form method="post">
        <label for="number">Enter the value of n:</label>
        <input type="number" id="number" name="number" min="1" required>
        <button type="submit">Generate</button>
      </form>
      {% if sequence %}
      <div class="result">
        <h2>Fibonacci Sequence:</h2>
        <p>{{ sequence|safe }}</p>
      </div>
      {% endif %}
    </div>
  </body>
</html>
'''

@app.route('/', methods=['GET', 'POST'])
def index():
    sequence = None
    if request.method == 'POST':
        n = int(request.form['number'])
        sequence = generate_fibonacci(n)
    return render_template_string(html_template, sequence=sequence)

def generate_fibonacci(n):
    prev = 0
    curr = 1
    fibonacci_sequence = [prev, curr]
    for _ in range(n - 2):
        next_num = curr + prev
        prev = curr
        curr = next_num
        fibonacci_sequence.append(next_num)
    return ' '.join(map(str, fibonacci_sequence))

if __name__ == '__main__':
    app.run()
```
